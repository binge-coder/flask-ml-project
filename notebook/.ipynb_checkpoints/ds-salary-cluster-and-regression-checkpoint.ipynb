```python
def plot_feature_importance1(importance,names,model_type):
#Create arrays from feature importance and feature names
    feature_importance = np.array(importance)
    feature_names = np.array(names)
    #Create a DataFrame using a Dictionary
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)
    #Sort the DataFrame in order decreasing feature importance
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)
    #Define size of bar plot
    plt.figure(figsize=(10,20))
    #Plot Searborn bar chart
    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])
    #Add chart labels
    plt.title(model_type + 'FEATURE IMPORTANCE')
    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')
```

```python
dfmodel['company types'] = hc_model.labels_
dfmodel
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import pairwise_distances
import numpy as np

def calculate_wcss(X, n_clusters):
    # Perform Agglomerative Clustering
    model = AgglomerativeClustering(n_clusters=n_clusters)
    model = model.fit(X)
    labels = model.labels_

    # Calculate centroids for each cluster
    centroids = np.array([X[labels == i].mean(axis=0) for i in range(n_clusters)])

    # Calculate squared distances for each point to its cluster centroid
    squared_distances = pairwise_distances(X, centroids, metric='euclidean')**2

    # Calculate WCSS by summing the minimum squared distance for each point to its centroid
    wcss = np.sum([squared_distances[i, label] for i, label in enumerate(labels)])
    
    return wcss
    
Sum_of_squared_distances=[]
K=np.arange(1,15)
for i in K:
    hc_model=AgglomerativeClustering(n_clusters=i, affinity='euclidean', linkage='ward')
    hc_model=hc_model.fit(df_cluster)
    Sum_of_squared_distances.append(calculate_wcss(df_cluster, i))  # kmeans.inertia_ is WCSSQ
plt.figure(figsize=(6,6))
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method to find Optimal k')
plt.show()  
```
